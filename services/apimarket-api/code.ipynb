````
import json
import re
from bs4 import BeautifulSoup
from metagpt.tools.libs.web_scraping import view_page_element_to_scrape

# Función para analizar la estructura de la página actual
async def analizar_estructura_eqhuma():
    # Obtener el contenido HTML de la página para análisis
    html_content = await view_page_element_to_scrape(
        url="https://eqhuma.mx/marketplace",
        requirement="Analizar la estructura del marketplace, categorías de APIs disponibles y características",
        keep_links=True
    )
    
    # Procesar el HTML con BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 1. Estructura general del marketplace
    estructura = {
        "título_principal": soup.find('h1').text.strip() if soup.find('h1') else "No encontrado",
        "secciones_principales": [],
        "categorías": []
    }
    
    # Buscar secciones principales
    secciones = soup.find_all('h2')
    for seccion in secciones:
        estructura["secciones_principales"].append(seccion.text.strip())
    
    # 2. Tipos de APIs disponibles
    apis_disponibles = []
    api_cards = soup.select('.card, .api-card, [class*="api"]')
    
    for card in api_cards:
        api_info = {}
        
        # Intentar extraer título de API
        titulo = card.select_one('h3, h4, .title, .api-title')
        if titulo:
            api_info["nombre"] = titulo.text.strip()
        
        # Intentar extraer descripción
        descripcion = card.select_one('.description, p, .api-description')
        if descripcion:
            api_info["descripcion"] = descripcion.text.strip()
            
        # Intentar extraer categoría
        categoria = card.select_one('.category, .tag, .api-category')
        if categoria:
            api_info["categoria"] = categoria.text.strip()
            
        # Intentar extraer proveedor
        proveedor = card.select_one('.provider, .vendor, .api-provider')
        if proveedor:
            api_info["proveedor"] = proveedor.text.strip()
            
        if api_info:
            apis_disponibles.append(api_info)
    
    # 3. Elementos de filtrado y búsqueda
    elementos_busqueda = {
        "barra_busqueda": bool(soup.select_one('input[type="search"], .search-input')),
        "filtros": []
    }
    
    # Identificar filtros
    filtros = soup.select('.filter, .filter-option, [class*="filter"]')
    for filtro in filtros:
        elementos_busqueda["filtros"].append(filtro.text.strip())
    
    # 4. Mecanismos de autenticación visibles
    auth_elementos = {
        "login_visible": bool(soup.select_one('.login, .signin, [href*="login"], [href*="signin"]')),
        "registro_visible": bool(soup.select_one('.register, .signup, [href*="register"], [href*="signup"]')),
        "perfil_usuario": bool(soup.select_one('.profile, .user-profile, .avatar'))
    }
    
    # Resultados consolidados
    resultados = {
        "estructura_general": estructura,
        "apis_disponibles": apis_disponibles,
        "elementos_busqueda": elementos_busqueda,
        "autenticacion": auth_elementos
    }
    
    return resultados

# Ejecutar el análisis
resultado_analisis = await analizar_estructura_eqhuma()

# Imprimir el análisis con formato
print("ANÁLISIS DE LA PLATAFORMA eqhuma.MX")
print("======================================\n")

print("1. ESTRUCTURA GENERAL DEL MARKETPLACE")
print("-" * 40)
print(f"Título principal: {resultado_analisis['estructura_general']['título_principal']}")
print("Secciones principales:")
for seccion in resultado_analisis['estructura_general']['secciones_principales']:
    print(f"- {seccion}")
print("\n")

print("2. APIS DISPONIBLES")
print("-" * 40)
if resultado_analisis['apis_disponibles']:
    for i, api in enumerate(resultado_analisis['apis_disponibles'], 1):
        print(f"API {i}:")
        for key, value in api.items():
            print(f"  {key.capitalize()}: {value}")
        print()
else:
    print("No se pudieron identificar claramente las APIs en el formato actual de la página.")
print("\n")

print("3. ELEMENTOS DE BÚSQUEDA Y FILTRADO")
print("-" * 40)
print(f"Barra de búsqueda: {'Disponible' if resultado_analisis['elementos_busqueda']['barra_busqueda'] else 'No encontrada'}")
print("Filtros identificados:")
if resultado_analisis['elementos_busqueda']['filtros']:
    for filtro in resultado_analisis['elementos_busqueda']['filtros']:
        print(f"- {filtro}")
else:
    print("- No se detectaron filtros específicos")
print("\n")

print("4. MECANISMOS DE AUTENTICACIÓN")
print("-" * 40)
print(f"Login visible: {'Sí' if resultado_analisis['autenticacion']['login_visible'] else 'No'}")
print(f"Registro visible: {'Sí' if resultado_analisis['autenticacion']['registro_visible'] else 'No'}")
print(f"Perfil de usuario: {'Visible' if resultado_analisis['autenticacion']['perfil_usuario'] else 'No visible'}")
````

````
import asyncio
from bs4 import BeautifulSoup
from metagpt.tools.libs.web_scraping import view_page_element_to_scrape
from metagpt.tools.libs.editor import Editor

async def analizar_apis_disponibles():
    """
    Analiza en detalle las APIs disponibles en eqhuma, extrayendo categorías,
    nombres, descripciones y otros detalles relevantes.
    """
    print("ANÁLISIS DETALLADO DE APIS EN eqhuma.MX")
    print("==========================================\n")
    
    # Obtener contenido HTML de la página para análisis más detallado
    html_content = await view_page_element_to_scrape(
        url="https://eqhuma.mx/marketplace",
        requirement="Extraer información detallada sobre las APIs disponibles, incluyendo categorías como SAT, IMSS, RENAPO, nombres, descripciones y detalles relevantes",
        keep_links=True
    )
    
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 1. Identificar categorías específicas de API (SAT, IMSS, RENAPO, etc.)
    categorias_encontradas = set()
    elementos_categoria = soup.select('.category, .badge, .tag, a[href*="category"], h3, h4')
    
    for elemento in elementos_categoria:
        texto = elemento.text.strip()
        # Buscar categorías mencionadas en el texto
        categorias_potenciales = ['SAT', 'IMSS', 'RENAPO', 'RFC', 'CURP', 'INE', 'INFONAVIT', 'SHCP']
        for categoria in categorias_potenciales:
            if categoria in texto:
                categorias_encontradas.add(categoria)
    
    # 2. Buscar secciones de API más estructuradas
    secciones_api = []
    card_elements = soup.select('.card, [class*="card"], .api-item, [class*="product"], article, [class*="service"]')
    
    for card in card_elements:
        api_info = {}
        
        # Extraer título
        title_elem = card.select_one('h2, h3, h4, .title, [class*="title"], [class*="heading"], strong')
        if title_elem:
            api_info['nombre'] = title_elem.text.strip()
        
        # Extraer descripción
        desc_elem = card.select_one('p, .description, [class*="desc"], [class*="content"]')
        if desc_elem:
            api_info['descripcion'] = desc_elem.text.strip()
        
        # Intentar identificar categoría
        category_elem = card.select_one('.category, .badge, .tag, [class*="category"], [class*="tag"]')
        if category_elem:
            api_info['categoria'] = category_elem.text.strip()
        
        # Intentar identificar precio o plan
        price_elem = card.select_one('[class*="price"], [class*="cost"], [class*="plan"]')
        if price_elem:
            api_info['precio'] = price_elem.text.strip()
        
        # Verificar si hay botones de acción
        action_buttons = card.select('button, .button, [class*="btn"], a.cta')
        if action_buttons:
            api_info['acciones'] = [btn.text.strip() for btn in action_buttons if btn.text.strip()]
        
        if api_info.get('nombre') or api_info.get('descripcion'):
            secciones_api.append(api_info)
    
    # 3. Analizar enlaces relacionados con APIs específicas
    enlaces_api = []
    links = soup.select('a[href*="api"], a[href*="product"], a[href*="service"]')
    
    for link in links:
        if link.text.strip():
            # Verificar si es una API específica
            href = link.get('href', '')
            api_name = link.text.strip()
            
            # Comprobar si es una API específica
            for categoria in categorias_potenciales:
                if categoria in api_name or categoria in href:
                    enlaces_api.append({
                        'nombre': api_name,
                        'enlace': href
                    })
                    break
    
    # 4. Buscar secciones destacadas o carruseles de API
    secciones_destacadas = {}
    secciones = soup.select('section, [class*="section"], [class*="featured"], [class*="popular"]')
    
    for seccion in secciones:
        titulo_seccion = seccion.select_one('h2, h3, .heading, [class*="title"]')
        if titulo_seccion:
            nombre_seccion = titulo_seccion.text.strip()
            apis_en_seccion = []
            
            # Buscar APIs dentro de esta sección
            api_cards = seccion.select('.card, [class*="item"], [class*="api"], [class*="product"]')
            for card in api_cards:
                nombre_api = card.select_one('h3, h4, .title')
                if nombre_api:
                    apis_en_seccion.append(nombre_api.text.strip())
            
            if apis_en_seccion:
                secciones_destacadas[nombre_seccion] = apis_en_seccion
    
    # 5. Identificar elementos de documentación y recursos
    documentacion = {
        'enlaces_docs': [],
        'recursos': []
    }
    
    # Enlaces de documentación
    doc_links = soup.select('a[href*="doc"], a[href*="docs"], a[href*="documentation"], a[href*="swagger"]')
    for link in doc_links:
        if link.text.strip():
            documentacion['enlaces_docs'].append({
                'nombre': link.text.strip(),
                'enlace': link.get('href', '')
            })
    
    # Enlaces de recursos
    recurso_links = soup.select('a[href*="resource"], a[href*="support"], a[href*="help"]')
    for link in recurso_links:
        if link.text.strip():
            documentacion['recursos'].append({
                'nombre': link.text.strip(),
                'enlace': link.get('href', '')
            })
    
    # === Mostrar resultados ===
    print("1. CATEGORÍAS DE APIs IDENTIFICADAS")
    print("-" * 40)
    if categorias_encontradas:
        for categoria in sorted(categorias_encontradas):
            print(f"- {categoria}")
    else:
        print("No se identificaron categorías específicas en el formato actual.")
    print("\n")
    
    print("2. APIS DISPONIBLES DETALLADAS")
    print("-" * 40)
    if secciones_api:
        for i, api in enumerate(secciones_api, 1):
            print(f"API {i}: {api.get('nombre', 'Sin nombre')}")
            for key, value in api.items():
                if key != 'nombre':
                    print(f"  - {key.capitalize()}: {value}")
            print()
    else:
        print("No se identificaron secciones de API estructuradas.")
    print("\n")
    
    print("3. ENLACES DE API ESPECÍFICOS")
    print("-" * 40)
    if enlaces_api:
        for i, enlace in enumerate(enlaces_api, 1):
            print(f"{i}. {enlace['nombre']}")
            print(f"   Enlace: {enlace['enlace']}")
    else:
        print("No se identificaron enlaces específicos a APIs individuales.")
    print("\n")
    
    print("4. SECCIONES DESTACADAS DE APIs")
    print("-" * 40)
    if secciones_destacadas:
        for seccion, apis in secciones_destacadas.items():
            print(f"Sección: {seccion}")
            for api in apis:
                print(f"  - {api}")
            print()
    else:
        print("No se identificaron secciones destacadas de APIs.")
    print("\n")
    
    print("5. DOCUMENTACIÓN Y RECURSOS")
    print("-" * 40)
    if documentacion['enlaces_docs']:
        print("Enlaces de documentación:")
        for doc in documentacion['enlaces_docs']:
            print(f"  - {doc['nombre']} ({doc['enlace']})")
    else:
        print("No se identificaron enlaces de documentación específicos.")
        
    if documentacion['recursos']:
        print("\nRecursos adicionales:")
        for recurso in documentacion['recursos']:
            print(f"  - {recurso['nombre']} ({recurso['enlace']})")
    else:
        print("\nNo se identificaron recursos adicionales.")
    
    # Crear diccionario estructurado con toda la información
    return {
        "categorias": sorted(list(categorias_encontradas)),
        "apis_disponibles": secciones_api,
        "enlaces_api": enlaces_api,
        "secciones_destacadas": secciones_destacadas,
        "documentacion": documentacion
    }

# Ejecutar el análisis detallado
analisis_apis = await analizar_apis_disponibles()

# Para facilitar el análisis posterior, guarda los resultados en un archivo
resultados_estructurados = {
    "fecha_analisis": "2023-07-29",
    "plataforma": "eqhuma.mx",
    "resultados": analisis_apis
}

# Crear directorio para guardar resultados si no existe
try:
    await asyncio.to_thread(lambda: __import__('os').makedirs("eqhuma_analisis", exist_ok=True))
    
    # Guardar resultados en un archivo JSON para análisis posterior
    import json
    with open("eqhuma_analisis/resultados_apis.json", "w", encoding="utf-8") as f:
        json.dump(resultados_estructurados, f, ensure_ascii=False, indent=2)
    
    print("\nLos resultados han sido guardados en 'eqhuma_analisis/resultados_apis.json'")
except Exception as e:
    print(f"No se pudieron guardar los resultados: {e}")
````

````
import asyncio
from bs4 import BeautifulSoup
from metagpt.tools.libs.web_scraping import view_page_element_to_scrape

async def analizar_api_especifica():
    """
    Analiza en detalle la API específica de INFONAVIT para comprender la estructura
    de presentación de APIs en la plataforma eqhuma.
    """
    print("ANÁLISIS DETALLADO DE API ESPECÍFICA: BUSCAR CRÉDITO INFONAVIT")
    print("============================================================\n")
    
    # Obtener contenido HTML de la página de API específica
    html_content = await view_page_element_to_scrape(
        url="https://eqhuma.mx/marketplace/buscar-credito-infonavit",
        requirement="Extraer información detallada sobre la estructura de la API, incluyendo documentación, esquemas de entrada/salida, precios, y forma de presentación",
        keep_links=True
    )
    
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 1. Información general de la API
    api_info = {
        "nombre": "",
        "descripcion": "",
        "categoria": "",
        "proveedor": "",
        "tiempo_respuesta": "",
        "estadisticas_uso": ""
    }
    
    # Extraer nombre de la API
    nombre_elem = soup.select_one('h1, h2.api-title, .title h1, .api-header h1')
    if nombre_elem:
        api_info["nombre"] = nombre_elem.text.strip()
    
    # Extraer descripción
    desc_elem = soup.select_one('.api-description, .description, p.desc')
    if desc_elem:
        api_info["descripcion"] = desc_elem.text.strip()
    
    # Extraer categoría
    cat_elem = soup.select_one('.api-category, .category, .badge[data-category]')
    if cat_elem:
        api_info["categoria"] = cat_elem.text.strip()
    
    # Extraer proveedor si está disponible
    prov_elem = soup.select_one('.provider, .vendor, .publisher')
    if prov_elem:
        api_info["proveedor"] = prov_elem.text.strip()
    
    # Extraer tiempo de respuesta si está disponible
    tiempo_elem = soup.select_one('.response-time, .performance, [data-response-time]')
    if tiempo_elem:
        api_info["tiempo_respuesta"] = tiempo_elem.text.strip()
    
    # Extraer estadísticas de uso si están disponibles
    stats_elem = soup.select_one('.stats, .usage-stats, .api-stats')
    if stats_elem:
        api_info["estadisticas_uso"] = stats_elem.text.strip()
    
    # 2. Estructura de la documentación técnica
    documentacion_tecnica = {
        "endpoint": "",
        "metodo_http": "",
        "headers": [],
        "parametros_entrada": [],
        "ejemplo_respuesta": "",
        "codigos_respuesta": []
    }
    
    # Extraer endpoint y método HTTP
    endpoint_elem = soup.select_one('.endpoint, .api-url, code.url')
    if endpoint_elem:
        documentacion_tecnica["endpoint"] = endpoint_elem.text.strip()
    
    metodo_elem = soup.select_one('.http-method, .method, .verb')
    if metodo_elem:
        documentacion_tecnica["metodo_http"] = metodo_elem.text.strip()
    
    # Extraer parámetros de entrada
    param_elems = soup.select('.parameter, .param, .request-param')
    for param in param_elems:
        param_name = param.select_one('.param-name, .name')
        param_type = param.select_one('.param-type, .type')
        param_desc = param.select_one('.param-desc, .description')
        
        if param_name:
            param_info = {
                "nombre": param_name.text.strip(),
                "tipo": param_type.text.strip() if param_type else "No especificado",
                "descripcion": param_desc.text.strip() if param_desc else "No especificado"
            }
            documentacion_tecnica["parametros_entrada"].append(param_info)
    
    # Extraer ejemplo de respuesta
    resp_elem = soup.select_one('.response-example, .example, pre code, .json-response')
    if resp_elem:
        documentacion_tecnica["ejemplo_respuesta"] = resp_elem.text.strip()
    
    # Extraer códigos de respuesta
    status_elems = soup.select('.status-code, .response-code, .http-code')
    for status in status_elems:
        code = status.select_one('.code')
        desc = status.select_one('.code-description')
        
        if code:
            code_info = {
                "codigo": code.text.strip(),
                "descripcion": desc.text.strip() if desc else "No especificado"
            }
            documentacion_tecnica["codigos_respuesta"].append(code_info)
    
    # 3. Detalles de precios y planes
    precios = {
        "plan_basico": "",
        "plan_pro": "",
        "plan_enterprise": "",
        "modelo_precio": ""  # Por uso, suscripción, etc.
    }
    
    # Extraer información de precios
    precio_elems = soup.select('.pricing, .price, .plan, .subscription')
    for precio in precio_elems:
        plan_name = precio.select_one('.plan-name, .name')
        plan_price = precio.select_one('.plan-price, .price')
        
        if plan_name and plan_price:
            plan_text = plan_name.text.strip().lower()
            if "basic" in plan_text or "básico" in plan_text:
                precios["plan_basico"] = plan_price.text.strip()
            elif "pro" in plan_text or "profesional" in plan_text:
                precios["plan_pro"] = plan_price.text.strip()
            elif "enterprise" in plan_text or "empresa" in plan_text:
                precios["plan_enterprise"] = plan_price.text.strip()
    
    # Extraer modelo de precio si está disponible
    modelo_elem = soup.select_one('.pricing-model, .billing-type')
    if modelo_elem:
        precios["modelo_precio"] = modelo_elem.text.strip()
    
    # 4. Elementos de UI y presentación
    ui_elementos = {
        "pestañas_documentacion": [],
        "botones_accion": [],
        "widgets_interactivos": [],
        "indicadores_estado": []
    }
    
    # Identificar pestañas de documentación
    tab_elems = soup.select('.tab, .nav-tab, .doc-tab')
    for tab in tab_elems:
        if tab.text.strip():
            ui_elementos["pestañas_documentacion"].append(tab.text.strip())
    
    # Identificar botones de acción
    btn_elems = soup.select('button, .btn, .button')
    for btn in btn_elems:
        if btn.text.strip():
            ui_elementos["botones_accion"].append(btn.text.strip())
    
    # Identificar widgets interactivos
    widget_elems = soup.select('.widget, .interactive, .playground')
    for widget in widget_elems:
        widget_name = widget.get('id') or widget.get('class')
        if widget_name:
            ui_elementos["widgets_interactivos"].append(str(widget_name))
    
    # Identificar indicadores de estado
    status_ind_elems = soup.select('.status, .indicator, .health')
    for ind in status_ind_elems:
        if ind.text.strip():
            ui_elementos["indicadores_estado"].append(ind.text.strip())
    
    # 5. Elementos de integración
    integracion = {
        "codigo_ejemplo": {},
        "instrucciones_autenticacion": "",
        "documentacion_relacionada": []
    }
    
    # Extraer códigos de ejemplo en diferentes lenguajes
    code_blocks = soup.select('pre code, .code-example, .sample-code')
    for code in code_blocks:
        language = ""
        # Intentar determinar el lenguaje del código
        parent = code.parent
        if parent:
            lang_elem = parent.select_one('.language, [data-language]')
            if lang_elem:
                language = lang_elem.text.strip() or lang_elem.get('data-language', '')
        
        if not language:
            # Intentar inferir el lenguaje del código
            code_text = code.text.strip()
            if "import requests" in code_text:
                language = "python"
            elif "fetch(" in code_text:
                language = "javascript"
            elif "curl" in code_text:
                language = "curl"
            else:
                language = "unknown"
        
        integracion["codigo_ejemplo"][language] = code.text.strip()
    
    # Extraer instrucciones de autenticación
    auth_elem = soup.select_one('.authentication, .auth, .api-key')
    if auth_elem:
        integracion["instrucciones_autenticacion"] = auth_elem.text.strip()
    
    # Extraer enlaces a documentación relacionada
    doc_links = soup.select('a[href*="doc"], a[href*="documentation"], a[href*="guide"]')
    for link in doc_links:
        if link.text.strip():
            integracion["documentacion_relacionada"].append({
                "nombre": link.text.strip(),
                "enlace": link.get('href', '')
            })
    
    # === Mostrar resultados ===
    # 1. Información general de la API
    print("1. INFORMACIÓN GENERAL DE LA API")
    print("-" * 40)
    for key, value in api_info.items():
        if value:  # Solo mostrar campos con valor
            print(f"{key.replace('_', ' ').capitalize()}: {value}")
    print("\n")
    
    # 2. Estructura de la documentación técnica
    print("2. DOCUMENTACIÓN TÉCNICA")
    print("-" * 40)
    print(f"Endpoint: {documentacion_tecnica['endpoint']}")
    print(f"Método HTTP: {documentacion_tecnica['metodo_http']}")
    
    # Parámetros de entrada
    if documentacion_tecnica['parametros_entrada']:
        print("\nParámetros de entrada:")
        for i, param in enumerate(documentacion_tecnica['parametros_entrada'], 1):
            print(f"  {i}. {param['nombre']} ({param['tipo']}): {param['descripcion']}")
    else:
        print("\nParámetros de entrada: No especificados")
    
    # Ejemplo de respuesta
    if documentacion_tecnica['ejemplo_respuesta']:
        print("\nEjemplo de respuesta:")
        print(f"```\n{documentacion_tecnica['ejemplo_respuesta']}\n```")
    
    # Códigos de respuesta
    if documentacion_tecnica['codigos_respuesta']:
        print("\nCódigos de respuesta:")
        for code in documentacion_tecnica['codigos_respuesta']:
            print(f"  {code['codigo']}: {code['descripcion']}")
    print("\n")
    
    # 3. Detalles de precios y planes
    print("3. PRECIOS Y PLANES")
    print("-" * 40)
    has_price_info = False
    for key, value in precios.items():
        if value:
            has_price_info = True
            print(f"{key.replace('_', ' ').capitalize()}: {value}")
    
    if not has_price_info:
        print("Información de precios no especificada en la página actual")
    print("\n")
    
    # 4. Elementos de UI y presentación
    print("4. ELEMENTOS DE UI Y PRESENTACIÓN")
    print("-" * 40)
    
    if ui_elementos['pestañas_documentacion']:
        print("Pestañas de documentación:")
        for tab in ui_elementos['pestañas_documentacion']:
            print(f"  - {tab}")
    
    if ui_elementos['botones_accion']:
        print("\nBotones de acción:")
        for btn in ui_elementos['botones_accion']:
            print(f"  - {btn}")
    
    if ui_elementos['widgets_interactivos']:
        print("\nWidgets interactivos:")
        for widget in ui_elementos['widgets_interactivos']:
            print(f"  - {widget}")
    
    if ui_elementos['indicadores_estado']:
        print("\nIndicadores de estado:")
        for indicator in ui_elementos['indicadores_estado']:
            print(f"  - {indicator}")
    print("\n")
    
    # 5. Elementos de integración
    print("5. ELEMENTOS DE INTEGRACIÓN")
    print("-" * 40)
    
    if integracion['codigo_ejemplo']:
        print("Códigos de ejemplo:")
        for language, code in integracion['codigo_ejemplo'].items():
            print(f"\n{language.upper()}:")
            print(f"```\n{code}\n```")
    
    if integracion['instrucciones_autenticacion']:
        print("\nInstrucciones de autenticación:")
        print(integracion['instrucciones_autenticacion'])
    
    if integracion['documentacion_relacionada']:
        print("\nDocumentación relacionada:")
        for doc in integracion['documentacion_relacionada']:
            print(f"  - {doc['nombre']} ({doc['enlace']})")
    
    # Devolver resultados estructurados para análisis posterior
    resultados = {
        "informacion_general": api_info,
        "documentacion_tecnica": documentacion_tecnica,
        "precios": precios,
        "ui_elementos": ui_elementos,
        "integracion": integracion
    }
    
    return resultados

# Ejecutar el análisis de API específica
analisis_api = await analizar_api_especifica()

# Guardar resultados en formato JSON
try:
    import json
    import os
    
    # Crear directorio si no existe
    os.makedirs("eqhuma_analisis", exist_ok=True)
    
    # Guardar análisis en archivo JSON
    with open("eqhuma_analisis/analisis_api_infonavit.json", "w", encoding="utf-8") as f:
        json.dump(analisis_api, f, ensure_ascii=False, indent=2)
    
    print("\nLos resultados del análisis han sido guardados en 'eqhuma_analisis/analisis_api_infonavit.json'")
except Exception as e:
    print(f"No se pudieron guardar los resultados: {e}")
````

````
async def analizar_documentacion_tecnica_api():
    """
    Extrae información detallada sobre la documentación técnica de la API específica
    de INFONAVIT enfocándose en los parámetros de solicitud, estructura de respuesta,
    método de autenticación y otros detalles técnicos importantes.
    """
    print("ANÁLISIS DETALLADO DE DOCUMENTACIÓN TÉCNICA DE LA API")
    print("===================================================\n")
    
    # Obtener contenido HTML de la página de documentación de la API
    html_content = await view_page_element_to_scrape(
        url="https://eqhuma.mx/marketplace/buscar-credito-infonavit?activeTab=docs#?route=post-/api/infonavit/grupo/buscar-credito",
        requirement="Extraer los detalles técnicos de la API, incluyendo parámetros de solicitud, estructura de respuesta, autenticación y ejemplos de uso",
        keep_links=True
    )
    
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 1. Información de autenticación y endpoint
    autenticacion = {
        "tipo": "",
        "headers_requeridos": [],
        "proceso_obtencion": "",
    }
    
    endpoint_info = {
        "url_base": "",
        "ruta": "",
        "metodo_http": "",
        "descripcion_general": ""
    }
    
    # Extraer información de autenticación
    auth_elems = soup.select('.auth, .authentication, [class*="auth"], .security')
    for auth in auth_elems:
        auth_text = auth.text.strip()
        if auth_text:
            if "API Key" in auth_text or "apikey" in auth_text.lower():
                autenticacion["tipo"] = "API Key"
            elif "Bearer" in auth_text or "token" in auth_text.lower():
                autenticacion["tipo"] = "Bearer Token"
            elif "OAuth" in auth_text:
                autenticacion["tipo"] = "OAuth 2.0"
            else:
                autenticacion["tipo"] = "Custom"
            break
    
    # Extraer headers requeridos
    header_elems = soup.select('.headers, [class*="header"], .request-headers')
    for header in header_elems:
        header_rows = header.select('tr, .row, .param-row')
        for row in header_rows:
            name = row.select_one('.name, .header-name')
            value = row.select_one('.value, .header-value, .example')
            if name:
                header_info = {
                    "nombre": name.text.strip(),
                    "valor": value.text.strip() if value else "No especificado"
                }
                autenticacion["headers_requeridos"].append(header_info)
    
    # Extraer información del endpoint
    endpoint_elems = soup.select('.endpoint, .api-url, .request-url')
    for endpoint in endpoint_elems:
        endpoint_text = endpoint.text.strip()
        if endpoint_text:
            if "http" in endpoint_text:
                # Separar base URL y ruta
                import re
                match = re.search(r'(https?://[^/]+)(/.*)', endpoint_text)
                if match:
                    endpoint_info["url_base"] = match.group(1)
                    endpoint_info["ruta"] = match.group(2)
                else:
                    endpoint_info["ruta"] = endpoint_text
            else:
                endpoint_info["ruta"] = endpoint_text
    
    # Extraer método HTTP
    method_elems = soup.select('.http-method, .method, [class*="method"]')
    for method in method_elems:
        method_text = method.text.strip().upper()
        if method_text in ["GET", "POST", "PUT", "DELETE", "PATCH"]:
            endpoint_info["metodo_http"] = method_text
            break
    
    # Extraer descripción general
    desc_elems = soup.select('.description, .endpoint-description, .summary')
    for desc in desc_elems:
        desc_text = desc.text.strip()
        if desc_text and len(desc_text) > 5:  # Evitar textos vacíos o muy cortos
            endpoint_info["descripcion_general"] = desc_text
            break
    
    # 2. Parámetros de solicitud detallados
    parametros_solicitud = {
        "path_params": [],
        "query_params": [],
        "body_params": [],
        "example_request": ""
    }
    
    # Extraer parámetros de la ruta (path)
    path_param_elems = soup.select('.path-param, [class*="path-param"]')
    for param in path_param_elems:
        param_name = param.select_one('.name, .param-name')
        param_type = param.select_one('.type, .param-type')
        param_desc = param.select_one('.description, .param-description')
        param_req = param.select_one('.required, .is-required')
        
        if param_name:
            param_info = {
                "nombre": param_name.text.strip(),
                "tipo": param_type.text.strip() if param_type else "No especificado",
                "descripcion": param_desc.text.strip() if param_desc else "No especificado",
                "requerido": True if (param_req and "true" in param_req.text.lower()) else False
            }
            parametros_solicitud["path_params"].append(param_info)
    
    # Extraer parámetros de consulta (query)
    query_param_elems = soup.select('.query-param, [class*="query-param"]')
    for param in query_param_elems:
        param_name = param.select_one('.name, .param-name')
        param_type = param.select_one('.type, .param-type')
        param_desc = param.select_one('.description, .param-description')
        param_req = param.select_one('.required, .is-required')
        
        if param_name:
            param_info = {
                "nombre": param_name.text.strip(),
                "tipo": param_type.text.strip() if param_type else "No especificado",
                "descripcion": param_desc.text.strip() if param_desc else "No especificado",
                "requerido": True if (param_req and "true" in param_req.text.lower()) else False
            }
            parametros_solicitud["query_params"].append(param_info)
    
    # Extraer parámetros del cuerpo (body)
    body_param_elems = soup.select('.body-param, [class*="body-param"], .request-body, .parameter-row')
    for param in body_param_elems:
        param_name = param.select_one('.name, .param-name')
        param_type = param.select_one('.type, .param-type')
        param_desc = param.select_one('.description, .param-description')
        param_req = param.select_one('.required, .is-required')
        
        if param_name:
            param_info = {
                "nombre": param_name.text.strip(),
                "tipo": param_type.text.strip() if param_type else "No especificado",
                "descripcion": param_desc.text.strip() if param_desc else "No especificado",
                "requerido": True if (param_req and "true" in param_req.text.lower()) else False
            }
            parametros_solicitud["body_params"].append(param_info)
    
    # Extraer ejemplo de solicitud
    req_example_elems = soup.select('.request-example, pre.request, [class*="request-example"]')
    for example in req_example_elems:
        example_text = example.text.strip()
        if example_text:
            parametros_solicitud["example_request"] = example_text
            break
    
    # 3. Estructura de respuesta
    estructura_respuesta = {
        "esquema": [],
        "ejemplo_respuesta": "",
        "codigos_respuesta": []
    }
    
    # Extraer esquema de respuesta
    schema_elems = soup.select('.response-schema, [class*="response-schema"], .response-model')
    for schema in schema_elems:
        schema_rows = schema.select('tr, .row, .schema-row')
        for row in schema_rows:
            prop_name = row.select_one('.property, .name')
            prop_type = row.select_one('.type, .property-type')
            prop_desc = row.select_one('.description, .property-description')
            
            if prop_name:
                prop_info = {
                    "propiedad": prop_name.text.strip(),
                    "tipo": prop_type.text.strip() if prop_type else "No especificado",
                    "descripcion": prop_desc.text.strip() if prop_desc else "No especificado"
                }
                estructura_respuesta["esquema"].append(prop_info)
    
    # Extraer ejemplo de respuesta
    resp_example_elems = soup.select('.response-example, pre.response, [class*="response-example"]')
    for example in resp_example_elems:
        example_text = example.text.strip()
        if example_text:
            estructura_respuesta["ejemplo_respuesta"] = example_text
            break
    
    # Extraer códigos de respuesta
    status_elems = soup.select('.status-code, .response-code, .http-status, [class*="status-code"]')
    for status in status_elems:
        code = status.select_one('.code, .status')
        desc = status.select_one('.description, .message')
        
        if code:
            code_info = {
                "codigo": code.text.strip(),
                "descripcion": desc.text.strip() if desc else "No especificado"
            }
            estructura_respuesta["codigos_respuesta"].append(code_info)
    
    # 4. Ejemplos de integración
    ejemplos_integracion = {
        "lenguajes": {}
    }
    
    # Extraer ejemplos de código en diferentes lenguajes
    code_blocks = soup.select('pre code, .code-example, .sample-code, [class*="code-sample"]')
    for code in code_blocks:
        # Intentar determinar el lenguaje
        language = "desconocido"
        parent = code.parent
        
        # Buscar indicadores de lenguaje
        if parent:
            lang_indicator = parent.get('class', [])
            lang_text = ' '.join(lang_indicator).lower()
            
            if any(lang in lang_text for lang in ['python', 'py']):
                language = 'python'
            elif any(lang in lang_text for lang in ['javascript', 'js']):
                language = 'javascript'
            elif any(lang in lang_text for lang in ['curl', 'bash', 'shell']):
                language = 'curl'
            elif any(lang in lang_text for lang in ['java']):
                language = 'java'
            elif any(lang in lang_text for lang in ['csharp', 'c#', 'dotnet']):
                language = 'csharp'
            elif any(lang in lang_text for lang in ['ruby']):
                language = 'ruby'
            
        # Si no se encontró clase, intentar inferir del contenido
        if language == "desconocido":
            code_text = code.text.lower()
            if "import requests" in code_text:
                language = "python"
            elif "fetch(" in code_text or "xhr" in code_text:
                language = "javascript"
            elif "curl" in code_text:
                language = "curl"
            elif "system.net.webclient" in code_text:
                language = "csharp"
        
        # Guardar el ejemplo de código
        if language not in ejemplos_integracion["lenguajes"]:
            ejemplos_integracion["lenguajes"][language] = []
        
        ejemplos_integracion["lenguajes"][language].append(code.text.strip())
    
    # === Mostrar resultados ===
    # 1. Información de autenticación y endpoint
    print("1. AUTENTICACIÓN Y ENDPOINT")
    print("-" * 40)
    print(f"Tipo de autenticación: {autenticacion['tipo'] or 'No especificado'}")
    
    if autenticacion['headers_requeridos']:
        print("\nHeaders requeridos:")
        for header in autenticacion['headers_requeridos']:
            print(f"  - {header['nombre']}: {header['valor']}")
    else:
        print("\nHeaders requeridos: No especificados")
    
    print(f"\nURL base: {endpoint_info['url_base'] or 'No especificada'}")
    print(f"Ruta: {endpoint_info['ruta'] or 'No especificada'}")
    print(f"Método HTTP: {endpoint_info['metodo_http'] or 'No especificado'}")
    
    if endpoint_info['descripcion_general']:
        print(f"\nDescripción general: {endpoint_info['descripcion_general']}")
    print("\n")
    
    # 2. Parámetros de solicitud
    print("2. PARÁMETROS DE SOLICITUD")
    print("-" * 40)
    
    if parametros_solicitud['path_params']:
        print("Parámetros de ruta (path):")
        for param in parametros_solicitud['path_params']:
            req_text = "Requerido" if param['requerido'] else "Opcional"
            print(f"  - {param['nombre']} ({param['tipo']}) [{req_text}]: {param['descripcion']}")
    
    if parametros_solicitud['query_params']:
        print("\nParámetros de consulta (query):")
        for param in parametros_solicitud['query_params']:
            req_text = "Requerido" if param['requerido'] else "Opcional"
            print(f"  - {param['nombre']} ({param['tipo']}) [{req_text}]: {param['descripcion']}")
    
    if parametros_solicitud['body_params']:
        print("\nParámetros del cuerpo (body):")
        for param in parametros_solicitud['body_params']:
            req_text = "Requerido" if param['requerido'] else "Opcional"
            print(f"  - {param['nombre']} ({param['tipo']}) [{req_text}]: {param['descripcion']}")
    
    if parametros_solicitud['example_request']:
        print("\nEjemplo de solicitud:")
        print(f"```\n{parametros_solicitud['example_request']}\n```")
    else:
        print("\nNo se encontró ejemplo de solicitud")
    print("\n")
    
    # 3. Estructura de respuesta
    print("3. ESTRUCTURA DE RESPUESTA")
    print("-" * 40)
    
    if estructura_respuesta['esquema']:
        print("Esquema de respuesta:")
        for prop in estructura_respuesta['esquema']:
            print(f"  - {prop['propiedad']} ({prop['tipo']}): {prop['descripcion']}")
    else:
        print("No se encontró esquema de respuesta detallado")
    
    if estructura_respuesta['ejemplo_respuesta']:
        print("\nEjemplo de respuesta:")
        print(f"```\n{estructura_respuesta['ejemplo_respuesta']}\n```")
    else:
        print("\nNo se encontró ejemplo de respuesta")
    
    if estructura_respuesta['codigos_respuesta']:
        print("\nCódigos de respuesta:")
        for code in estructura_respuesta['codigos_respuesta']:
            print(f"  - {code['codigo']}: {code['descripcion']}")
    print("\n")
    
    # 4. Ejemplos de integración
    print("4. EJEMPLOS DE INTEGRACIÓN")
    print("-" * 40)
    
    if ejemplos_integracion['lenguajes']:
        for language, examples in ejemplos_integracion['lenguajes'].items():
            print(f"\n{language.upper()}:")
            for i, example in enumerate(examples, 1):
                print(f"Ejemplo {i}:")
                print(f"```\n{example}\n```")
    else:
        print("No se encontraron ejemplos de integración")
    
    # Crear resultado estructurado para análisis posterior
    resultado = {
        "autenticacion": autenticacion,
        "endpoint_info": endpoint_info,
        "parametros_solicitud": parametros_solicitud,
        "estructura_respuesta": estructura_respuesta,
        "ejemplos_integracion": ejemplos_integracion
    }
    
    return resultado

# Ejecutar el análisis de documentación técnica
analisis_documentacion = await analizar_documentacion_tecnica_api()

# Guardar resultados en formato JSON
try:
    import json
    import os
    
    # Crear directorio si no existe
    os.makedirs("eqhuma_analisis", exist_ok=True)
    
    # Guardar análisis en archivo JSON
    with open("eqhuma_analisis/documentacion_tecnica_api.json", "w", encoding="utf-8") as f:
        json.dump(analisis_documentacion, f, ensure_ascii=False, indent=2)
    
    print("\nLos resultados del análisis de documentación técnica han sido guardados en 'eqhuma_analisis/documentacion_tecnica_api.json'")
except Exception as e:
    print(f"No se pudieron guardar los resultados: {e}")
````
